### 1.5 Tratamiento de los datos nulos

```python
porcentaje_nulos = df.isna().sum().sort_values(ascending=False) / len(df) * 100
print("Porcentaje de datos nulos por columna")
print(porcentaje_nulos)
```

```python
cols = df.isna().sum().sort_values(ascending=False)[:3].index
```

```python
msno.matrix(df[cols])
plt.show()
```

Tomando en consideración que los datos nulos presentes en el dataframe son bajos en consideración al resto de los datos, se procederá a realizar la imputación usando la técnica de KNN imputer.

#### 1.5.1 Imputación de los datos nulos

Separar los datos numericos de los datos categóricos

```python
numeric_df = df.select_dtypes(include='number')
```

```python
numeric_df.columns
```

Verificar la distribución de los datos antes de aplicar la imputación

```python
nrows, ncols = 1, 3
fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15,5))
sns.set_theme()


col_pos = 0
for idx, col in enumerate(numeric_df[cols].columns):

    # Visualización
    sns.histplot(data=numeric_df, x=numeric_df[col], ax=ax[col_pos], kde=True)
    ax[col_pos].set_title(f'Distribución de {col}')
    ax[col_pos].set_xlabel('Frecuencua')
    ax[col_pos].set_ylabel(col)

    # Controlar el número de la columna
    col_pos += 1

plt.tight_layout()
plt.show()
```
