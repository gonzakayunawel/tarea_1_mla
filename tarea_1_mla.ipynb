Tomando en consideración que los datos nulos presentes en el dataframe son bajos en consideración al resto de los datos, se procederá a realizar la imputación usando la técnica de KNN imputer.

#### 1.5.1 Imputación de los datos nulos

Separar los datos numericos de los datos categóricos

```python
numeric_df = df.select_dtypes(include='number')
```

```python
numeric_df.columns
```

Verificar la distribución de los datos antes de aplicar la imputación

```python
nrows, ncols = 1, 3
fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15,5))
sns.set_theme(style="darkgrid")


col_pos = 0
for idx, col in enumerate(numeric_df[cols].columns):

    # Visualización
    sns.histplot(data=numeric_df, x=numeric_df[col], ax=ax[col_pos], kde=True)
    ax[col_pos].set_title(f'Distribución de {col}')
    ax[col_pos].set_xlabel('Frecuencia')
    ax[col_pos].set_ylabel(col)

    # Controlar el número de la columna
    col_pos += 1

plt.tight_layout()
plt.show()
```

Realizar la imputación

```python
from sklearn.impute import KNNImputer

imputer = KNNImputer()

imputed_data = imputer.fit_transform(numeric_df)

df_imputed = pd.DataFrame(imputed_data, columns=numeric_df.columns)
```

Verificar la distribución de los datos después de la imputación

```python
df_data = df_imputed

nrows, ncols = 1, 3
fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15,5))
sns.set_theme(style="darkgrid")


col_pos = 0
for idx, col in enumerate(df_data[cols].columns):

    # Visualización
    sns.histplot(data=df_data, x=df_data[col], ax=ax[col_pos], kde=True)
    ax[col_pos].set_title(f'Distribución de {col}')
    ax[col_pos].set_xlabel('Frecuencia')
    ax[col_pos].set_ylabel(col)

    # Controlar el número de la columna
    col_pos += 1

plt.tight_layout()
plt.show()
```

Realizar la concatenación de los df despues de la imputación

```python
categoric_df = df.select_dtypes(exclude='number')
```

```python
df = pd.concat([categoric_df, df_imputed], axis=1)
```

### 1.8 Análisis de distribuciones, frecuencia y pruebas de normalidad de datos numéricos

```python
df_numeric_data = df.select_dtypes(include='number')
```

```python
df_numeric_data
```

```python
# Importamos la funcion probplot para graficar el Q-Q Plot
from scipy.stats import probplot

for col in df_numeric_data.columns:  # type: ignore

    plt.figure(figsize=(15, 4))

    if len(df_numeric_data[col].unique()) > 20:  # type: ignore
        plt.subplot(1, 3, 1)
        sns.kdeplot(data=df_numeric_data, x=col, fill=True, alpha=0.5, linewidth=2)  # type: ignore
        plt.xlabel(col)
        plt.ylabel("Densidad")
        plt.title(f"Densidad de {col}")
    else:
        plt.subplot(1, 3, 1)
        sns.histplot(df_numeric_data[col], bins=30, color="skyblue")  # type: ignore
        plt.xlabel(col)
        plt.ylabel("Frequencia")
        plt.title(f"Frecuencia de {col}")


    plt.subplot(1, 3, 2)
    sns.boxplot(data=df_numeric_data[col], color="darkblue")  # type: ignore
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.title(f"Boxplot de {col}")

    plt.subplot(1, 3, 3)
    probplot(x=df_numeric_data[col], dist="norm", plot=plt)  # type: ignore
    plt.xlabel("Cuantiles Teóricos")
    plt.ylabel("Cuantiles Observados")
    plt.title(f"QQplot de {col}")


    plt.tight_layout()

    plt.show()
```

#### 1.8.1 Interpretación estadística y consideracions para el tratamiento de outliers

La variable "mora" es binaria, con una distribución altamente desequilibrada a favor de los morosos (1). Hay una minoría significativa de casos no morosos (0), lo que sugiere que la tasa de morosidad es relativamente alta en esta población de clientes bancarios.

La variable "atraso" muestra una distribución altamente sesgada a la derecha, con la mayoría de los clientes teniendo pocos o ningún día de atraso. Sin embargo, hay una cola larga que indica la presencia de casos extremos con atrasos significativos, llegando hasta aproximadamente 250 días en algunos casos excepcionales. Esto se expresa como valores atípicos tanto en el boxplot como en el qqplot de la variable.

La variable "edad" muestra una distribución aproximadamente normal con un ligero sesgo positivo, concentrándose la mayoría de los clientes entre los 30 y 50 años. Se observan algunos valores atípicos en edades más avanzadas, pero en general la distribución sigue bastante de cerca una curva normal. Para este caso se eliminará de los registros a las personas con una edad mayor a 80 años ya que a esa edad la capacidad de generar mayores ingresos que la jubilación se encuenta limitada.

La variable "dias_lab" (días laborados) presenta una distribución fuertemente sesgada a la derecha, con la mayoría de los empleados concentrados en periodos más cortos de empleo. Se observan valores atípicos significativos, indicando algunos empleados con periodos de trabajo excepcionalmente largos, llegando hasta aproximadamente 20,000 días (equivalente a más de 54 años). Para este caso se hará una eliminación de outliers basado en el hecho de que si se comienza a trabajar a las 18 y se jubila los 65 solo se podría trabajar 47 años.

La variable "exp_sf" muestra una distribución multimodal con un pico principal en valores bajos y picos secundarios alrededor de los 75-100 meses. La distribución es asimétrica hacia la derecha, indicando que mientras la mayoría de los clientes tienen relativamente poca experiencia en el sistema financiero, existe un grupo significativo con experiencia de mediana a larga duración.

La variable "nivel_ahorro" muestra una distribución altamente asimétrica, con una concentración extrema en los valores más bajos (0) y otra concentración significativa en el valor más alto (12). Esto sugiere una polarización en los hábitos de ahorro de los clientes, con la mayoría teniendo poco o ningún ahorro, y un grupo menor pero notable con altos niveles de ahorro. Esta variable parece tener un comportamiento más de tipo categórico que numérico.

La variable "ingreso" presenta una distribución fuertemente sesgada a la derecha, con la mayoría de los clientes concentrados en niveles de ingresos más bajos. Se observa una larga cola hacia la derecha, indicando la presencia de algunos clientes con ingresos significativamente más altos que la media. Esta distribución sugiere una desigualdad considerable en los ingresos de la clientela bancaria.

La variable "linea_sf" muestra una distribución muy sesgada a la derecha, con la mayoría de los clientes teniendo líneas de crédito relativamente bajas. Se observa una larga cola hacia la derecha, indicando que un pequeño número de clientes tiene líneas de crédito excepcionalmente altas. Esta distribución sugiere una gran disparidad en las líneas de crédito asignadas, posiblemente correlacionada con los niveles de ingresos o la solvencia crediticia de los clientes.

La variable "deuda_sf" presenta una distribución fuertemente sesgada a la derecha, con la mayoría de los clientes concentrados en niveles de deuda bajos. Se observa una larga cola hacia la derecha y un pequeño pico secundario en valores más altos, indicando que mientras la mayoría tiene deudas relativamente pequeñas, existe un grupo de clientes con niveles de deuda significativamente mayores.

La variable "score" presenta una distribución bimodal, con dos picos principales alrededor de 200 y 225, lo que sugiere que la cartera de clientes incluye una variedad de perfiles crediticios, donde algunos tienen puntajes altos y otros más bajos. Por último, el QQplot muestra que la distribución de los puntajes se ajusta bien a una distribución normal, aunque hay algunos valores atípicos en los extremos.

La variable "clasif_sbs" presenta una frecuencia donde la mayoría de los valores se agrupan entre 0 y 1, con algunos valores alcanzando hasta 4, lo que indica una distribución asimétrica y sesgada hacia los valores más bajos. El boxplot muestra una mediana cercana a 1, junto con algunos valores atípicos por encima de 3, lo que respalda la asimetría observada en el histograma. Además, el QQplot sugiere que la distribución de clasif_sbs se desvía notablemente de una distribución normal, evidenciando una clara tendencia asimétrica y sesgada hacia los valores bajos.
