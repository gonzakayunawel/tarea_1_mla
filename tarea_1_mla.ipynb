### 1.8.3 Aplicación de filtros de outliers

#### 1.8.3.1 Filtrar outliers en columnas usando IQR

```python
df = df.copy()

def filter_outliers_iqr(df, column):
    # Calcular el primer cuartil (Q1) y el tercer cuartil (Q3)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Definir los límites inferior y superior
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Generar una máscara booleana para filtrar outliers
    mask = (df[column] >= lower_bound) & (df[column] <= upper_bound)

    # Aplicar la máscara al DataFrame
    df.loc[~mask, column] = np.nan
    return df

columns_to_filter = ['atraso', 'exp_sf', 'ingreso', 'linea_sf', 'deuda_sf', 'score']

# Aplicar el filtro de outliers IQR a las columnas especificadas
for col in columns_to_filter:
    df = filter_outliers_iqr(df, col)

df_cleaned = df.dropna()
```

#### 1.8.3.2 Filtro basado en los valores atípicos

```python
dias_lab_max = 47*365
# Eliminar valores superiores al límite en la columna 'dias_lab'
df_cleaned = df_cleaned[df_cleaned['dias_lab'] <= dias_lab_max]
```

## 2. Aplicar técnicas de clustering, como k-medias, k-medoides, dendrograma y DBSCAN, para identificar segmentos de clientes.

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn_extra.cluster import KMedoids
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score


# 1. Normalización/Estandarización de los datos numéricos
# Normalizamos los datos para que todas las características tengan una escala similar,
# ya que K-Means y otros algoritmos de clustering pueden ser sensibles a las diferencias de escala.
scaler = StandardScaler()
df_scaled = df.copy()  # Creamos una copia del dataframe original para no modificarlo directamente
df_scaled[df.select_dtypes(include=['float64']).columns] = scaler.fit_transform(df.select_dtypes(include=['float64']))

# 2. Codificación de Variables Categóricas
# Convertimos las columnas categóricas en variables numéricas utilizando one-hot encoding
df_encoded = pd.get_dummies(df_scaled, columns=['vivienda', 'zona', 'nivel_educ'])

# 3. K-Medias (K-Means)

# a. Determinar el número óptimo de clústeres usando el método del codo
# Creamos un gráfico para visualizar la inercia (suma de distancias dentro de los clústeres) en función del número de clústeres
inertia = []  # Lista para almacenar la inercia de cada número de clústeres
for k in range(1, 11):  # Probar con 1 a 10 clústeres
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_encoded)  # Ajustamos el modelo K-Means a los datos
    inertia.append(kmeans.inertia_)  # Almacenamos la inercia

# Graficamos el método del codo
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Número de clústeres')
plt.ylabel('Inercia')
plt.title('Método del Codo')
plt.show()

# b. Aplicar K-Medias con el número óptimo de clústeres
# Elige el número de clústeres basado en el gráfico del codo
n_clusters = 4  # Ejemplo, elige el número basado en el gráfico

kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_encoded['kmeans_cluster'] = kmeans.fit_predict(df_encoded)  # Predice y asigna los clústeres al dataframe

# 4. K-Medoides (PAM)

# Aplicamos K-Medoides, una variante de K-Means que es menos sensible a los valores atípicos
kmedoids = KMedoids(n_clusters=n_clusters, random_state=42)
df_encoded['kmedoids_cluster'] = kmedoids.fit_predict(df_encoded)  # Predice y asigna los clústeres al dataframe

# 5. Dendrograma (Clustering Jerárquico)

# a. Crear el dendrograma para visualizar la estructura de los clústeres
# El dendrograma nos ayuda a entender cómo se agrupan los datos en diferentes niveles
dendrogram = sch.dendrogram(sch.linkage(df_encoded, method='ward'))
plt.title('Dendrograma')
plt.xlabel('Ítems')
plt.ylabel('Distancia')
plt.show()
```

```python
# b. Aplicar clustering jerárquico usando el número de clústeres basado en el dendrograma
n_clusters = 4  # Ejemplo, elige el número basado en el dendrograma

hierarchical = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
df_encoded['hierarchical_cluster'] = hierarchical.fit_predict(df_encoded)  # Predice y asigna los clústeres al dataframe

# 6. DBSCAN

# Aplicamos DBSCAN, que es útil para detectar clústeres de forma arbitraria y manejar ruido
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Ajusta eps y min_samples según la densidad esperada de los clústeres
df_encoded['dbscan_cluster'] = dbscan.fit_predict(df_encoded)  # Predice y asigna los clústeres al dataframe

# 7. Evaluación de los Clústeres

# Evaluamos la calidad de los clústeres usando el índice de silueta, que mide qué tan similares son los puntos dentro de un clúster
# El índice de silueta varía de -1 a 1, donde valores más altos indican una mejor calidad del clustering

# Evaluar K-Medias
silhouette_kmeans = silhouette_score(df_encoded, df_encoded['kmeans_cluster'])
print(f'Silhouette Score for K-Means: {silhouette_kmeans}')

# Evaluar K-Medoides
silhouette_kmedoids = silhouette_score(df_encoded, df_encoded['kmedoids_cluster'])
print(f'Silhouette Score for K-Medoids: {silhouette_kmedoids}')

# Evaluar Clustering Jerárquico
silhouette_hierarchical = silhouette_score(df_encoded, df_encoded['hierarchical_cluster'])
print(f'Silhouette Score for Hierarchical: {silhouette_hierarchical}')

# Evaluar DBSCAN
# DBSCAN puede producir clústeres etiquetados como -1 (ruido), lo cual puede afectar el índice de silueta
# Filtrar el ruido
filtered_data = df_encoded[df_encoded['dbscan_cluster'] != -1]
filtered_labels = filtered_data['dbscan_cluster']

# Aseguramos que haya más de un clúster para evitar errores
if len(set(filtered_labels)) > 1:
    silhouette_dbscan = silhouette_score(filtered_data, filtered_labels)
    print(f'Silhouette Score for DBSCAN: {silhouette_dbscan}')
else:
    print("No se puede calcular el índice de silueta ya que no hay suficientes clústeres distintos después de filtrar el ruido.")
```

#### 2.1 Conclusión de Técnicas de Clustering

1. Comparación de Métodos de Clustering:
Basándonos en los índices de silueta obtenidos, podemos concluir que:

- Clustering Jerárquico: Este método obtuvo el mejor rendimiento con un índice de silueta de 0.2357. Un índice de silueta más alto indica que, en promedio, los puntos asignados a cada clúster están mejor agrupados y más claramente separados de otros clústeres.

- K-Means: El rendimiento fue intermedio con un índice de silueta de 0.2168. Aunque menor que el del clustering jerárquico, su índice sugiere una estructura de clúster moderada.

- K-Medoids: Presentó el rendimiento más bajo con un índice de silueta de 0.1565. Este resultado indica que los clústeres formados están menos definidos en comparación con los otros métodos.

2. Calidad del Clustering:
Las evaluaciones de los índices de silueta sugieren lo siguiente:

- Todos los índices son positivos pero relativamente bajos (en el rango de 0 a 0.25). Esto indica que, si bien hay cierta estructura de clúster, no es muy fuerte ni está claramente definida.

- Un índice de silueta cercano a 0 implica que los puntos están cerca de los límites decisionales entre clústeres, lo cual refleja que los clústeres no son tan separados y los puntos están dispersos.

3. Análisis del Dendrograma:
El dendrograma proporciona una visión jerárquica de la estructura de clúster que merece atención:

- Tres Grupos Principales: El dendrograma muestra tres grupos discernibles, marcados en colores naranja, verde y rojo. Esta visualización ayuda a identificar las relaciones jerárquicas y las distancias entre los clústeres.

- Altura de Conexiones: Las conexiones más altas en el dendrograma indican mayor disimilitud entre los grupos. Este detalle sugiere que los grupos formados presentan diferencias notables en ciertos puntos.

4. Limitaciones:
El análisis reveló ciertos desafíos y limitaciones:

- Método DBSCAN: No se pudo calcular el índice de silueta para DBSCAN debido a la falta de clústeres distintos una vez filtrado el ruido. Esto revela que DBSCAN no fue particularmente eficaz en identificar clústeres en este conjunto de datos específico.

Finalmente, es posible concluir que los resultados demuestran cierta presencia de estructura de clúster en los datos, aunque no demasiado pronunciada. Entre los métodos evaluados, el clustering jerárquico parece ser el más efectivo, aunque la calidad general del clustering es moderada.

## 4. Aplicar técnicas de reducción de dimensionalidad, como PCA, para visualizar los datos en un espacio de menor dimensionalidad y analizar su estructura

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Convertir variables categóricas en variables dummy
df_encoded = pd.get_dummies(df, columns=['vivienda', 'zona', 'nivel_educ'])

# Binarizar variables numéricas
# Convertimos cada variable numérica en una serie de columnas binarias basadas en un umbral
def binarize_column(column, threshold):
    return (column > threshold).astype(int)

# Aplicar la binarización a las columnas numéricas
for col in ['mora', 'atraso', 'edad', 'dias_lab', 'exp_sf', 'nivel_ahorro', 'ingreso', 'linea_sf', 'deuda_sf', 'score', 'clasif_sbs']:
    df_encoded[col] = binarize_column(df_encoded[col], threshold=df_encoded[col].median())

# Separar las características (X) para aplicar PCA
X = df_encoded.copy()

# Estandarizar los datos para que tengan media 0 y desviación estándar 1
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Aplicar PCA
# El número de componentes principales es un parámetro que puedes ajustar
pca = PCA(n_components=2)  # Reducción a 2 dimensiones para la visualización
X_pca = pca.fit_transform(X_scaled)

# Mostrar la varianza explicada por cada componente principal
print("Varianza explicada por cada componente:", pca.explained_variance_ratio_)
print("Varianza acumulada:", np.cumsum(pca.explained_variance_ratio_))
```

```python
# Crear un DataFrame para los componentes principales
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])

# Visualizar los datos en el espacio de los dos primeros componentes principales
plt.figure(figsize=(10, 7))
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.5)
plt.title('Visualización de Datos en el Espacio de PCA')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.grid(True)
plt.show()
```
